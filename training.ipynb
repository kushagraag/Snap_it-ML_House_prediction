### Removing NaN 

```python
df['PRODUCT_BRAND']=df['PRODUCT_BRAND'].fillna('missing')
df['PRODUCT_BRAND'].head(20)
```

```python
df['CATEGORY']=df['CATEGORY'].fillna(df['CATEGORY'].mode()[0])
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.head()
```

Dropping extra columns

```python
df=df.drop(['PRODUCT_ID', 'PRODUCT_NAME','PRODUCT_DESCRIPTION'], axis=1)
```

```python
df.head()
```

### Categorical

splitting "Category" 

```python
df[["Main_category", "Sub_category", "Sub_sub_category", "cat4", "cat5"]] = df.CATEGORY.str.split("/", expand=True)
df.head()
```

picking only 3 categories of each item

```python
df.drop(["CATEGORY", "cat4", "cat5"], axis = 1, inplace = True)
df.isna().sum()
df.head()
```

Main cateogry one hot encoding

```python
df = pd.get_dummies(df, columns=['Main_category'])
```

```python
df.head()
df.shape
```

Sub-category one-hot encoding (top 50) (approx half of original)

```python
sub_category_top50=[x for x in df.Sub_category.value_counts().sort_values(ascending=False).head(50).index]
sub_category_top50
```

```python
for i in sub_category_top50:
    df[i]=np.where(df['Sub_category']==i,1,0)
df.head()
```

Sub-sub-category one hot top 20

```python
sub_sub_category_top20=[x for x in df.Sub_sub_category.value_counts().sort_values(ascending=False).head(20).index]
sub_sub_category_top20
```

```python
for i in sub_sub_category_top20:
    df[i]=np.where(df['Sub_sub_category']==i,1,0)
df.head()
```

Product brand

```python
product_brand_top20=[x for x in df.PRODUCT_BRAND.value_counts().sort_values(ascending=False).head(20).index]
product_brand_top20
```

```python
for i in product_brand_top20:
    df[i]=np.where(df['PRODUCT_BRAND']==i,1,0)
df.head()
```
